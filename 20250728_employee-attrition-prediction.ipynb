```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("C:/Academic_work/work_forPy/学習用データ.csv")
train.head()
```

```python
test = pd.read_csv("C:/Academic_work/work_forPy/検証用データ.csv")
test.head()
```

```python
# データ型の確認
print(train.dtypes)
print(test.dtypes)
```

```python
train.info()
test.info()
```

```python
# 45歳以下に絞る（学習データ・検証データ両方）
train = train[train["年齢"] <= 45].reset_index(drop=True)
test = test[test["年齢"] <= 45].reset_index(drop=True)
```

```python
y_train = train['退職フラグ']
X_train = train.drop(columns = ["退職フラグ", "入社日", "退社日"], errors = 'ignore')
y_test = test['退職フラグ']
X_test = test.drop(columns=["退職フラグ", "入社日", "退社日"], errors = 'ignore')
```

```python
# get_dummies を改めて適用
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)

# テストデータを学習データに合わせる
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = sm.fit_resample(X_train, y_train)

# クラス数の確認
print("Before:", y_train.value_counts().to_dict())
print("After :", pd.Series(y_train_resampled).value_counts().to_dict())
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# 欠損値の確認
print(X_train.isnull().sum())
print(X_test.isnull().sum())
```

```python
from sklearn.metrics import classification_report
```

```python
# モデル構築
rf_model = RandomForestClassifier(max_depth=10, n_estimators=30, random_state=222)
rf_model.fit(X_train_resampled, y_train_resampled)
```

```python
# 予測
y_pred = rf_model.predict(X_test)

# 精度評価
print(classification_report(y_test, y_pred))
```

```python
# 評価用データで推測
pred = rf_model.predict(X_test) 
```

```python
# 正解率計算の関数読み込み
from sklearn.metrics import accuracy_score
# 正解率
accuracy_score(pred, y_test)
```

```python
# 混同行列作成の関数を読み込み
from sklearn.metrics import confusion_matrix
#混同行列で精度を確認する
labels = [1, 0] # ラベルの順序を指定
cm = confusion_matrix(y_test, pred, labels=labels) # 混同行列の取得&ラベル順序指定
columns_labels = ["pred_" + str(l) for l in labels]
index_labels = ["act_" + str(l) for l in labels]
cm = pd.DataFrame(cm, columns=columns_labels, index=index_labels)
cm
```

```python
# 分類レポート作成用の関数読み込み
from sklearn.metrics import classification_report
# 分類の評価レポート作成
print(classification_report(y_test, pred))
```

```python
# ROC曲線とAUC
# 予測確率を算出
prob_rf = rf_model.predict_proba(X_test)[:, 1] #分類の結果を確率で返す
# ROC曲線とAUC用の関数読み込み
from sklearn.metrics import roc_curve, roc_auc_score
# ROC曲線用データ作成
fpr, tpr, thresholds = roc_curve(y_test, prob_rf)
#内容の確認
print(fpr)
print(tpr)
print(thresholds)
```

```python
# ライブラリの読み込み
import matplotlib.pyplot as plt #グラフィックの描画
# ROC曲線描画
plt.plot(fpr, tpr, label='random forestss')
plt.fill_between(fpr, tpr, 0, alpha=0.1)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
# AUCの計算
roc_auc_score(y_test, prob_rf)
```

```python
# どういう値が計算されているか確認
# ROC曲線作成
fpr_all, tpr_all, th_all = roc_curve(y_test, prob_rf, drop_intermediate=False)
# データフレームに変換
df_roc = pd.DataFrame({'th_all': th_all, 'tpr_all': tpr_all, 'fpr_all': fpr_all})
print(df_roc)
```

```python
# 変数の重要度を取得
importances = rf_model.feature_importances_

# ラベルを X_train_resampled から取得する
label = X_train_resampled.columns

# 重要度を降順で並び替え
i = np.argsort(importances)[::-1]

# 上位10個だけ表示（任意）
for idx in i[:10]:
    print(f"{label[idx]}: {importances[idx]:.4f}")
```
